```python
#(Plus dur) Comment la pratique de plantation a-t-elle évolué au cours du temps ? Diversifie-t-on plus maintenant qu’avant? 
#clustering par tranche de 10 ans
# data[['ANNEEDEPLANTATION','ELEM_POINT_ID','GENRE_BOTA', 'ESPECE']].sort_values(by=['ELEM_POINT_ID'],ascending = True)
#compute population
data9=data[['ANNEEDEPLANTATION','ELEM_POINT_ID','GENRE_BOTA', 'ESPECE']].groupby(['ANNEEDEPLANTATION','GENRE_BOTA', 'ESPECE']).count()
data9=data9.rename(columns={'ELEM_POINT_ID':'POPULATION'}).reset_index()  	#get the columns back 
data9[['DECENIE']] = data9[['ANNEEDEPLANTATION']].map((lambda n : (n//10) * 10)) #create decades column
data9=data9[['GENRE_BOTA','ESPECE','POPULATION','DECENIE']] #remove annee column
data9=data9.groupby(['DECENIE','GENRE_BOTA','ESPECE']).sum()  #sum and groupeby by decades
data9=data9.reset_index().sort_values(by=['DECENIE','GENRE_BOTA','ESPECE'])
#compute proportion per decade 
sum_pop = data9.groupby(['DECENIE']).sum().rename({'POPULATION':'SUM_POP'}, axis=1).reset_index()
sum_pop = sum_pop[['DECENIE', 'SUM_POP']]
data_with_sumpop = data9.merge(sum_pop, how= 'left', on='DECENIE')
def funProp(r):
    return 100*r.iloc[0]/r.iloc[1]
data_with_sumpop.iloc[0,4]
data_with_sumpop[['POPULATION','SUM_POP']].apply(funProp,axis=1)
data_with_prop = data_with_sumpop.copy()
data_with_prop['PROP'] = data_with_sumpop[['POPULATION', 'SUM_POP']].apply(funProp,axis=1)
data_with_prop=data_with_prop[['DECENIE','GENRE_BOTA','ESPECE','PROP']]
data_with_prop
#Compute Standard deviation and plot
data_with_prop[['DECENIE','PROP']].groupby(['DECENIE']).std().rename(columns={'PROP':'STD_PROP'}).plot()
```
