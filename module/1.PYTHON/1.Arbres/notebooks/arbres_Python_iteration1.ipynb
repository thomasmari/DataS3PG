```python
#5. Extraire l’information concernant l’année de plantation pour l’ensemble des lignes (rows) 
#et stocker ces informations dans une autre variable de type List
#Nom des collonnes
with open('../data/trees.csv', newline='') as csvfile:
    tree_reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    i=0
    rows = []
    for row in tree_reader:
        rows.append(row)
    header = list(rows[0])
    n = len(header)
    print("Informations sur les collones:")
    print("\n".join([f"{i}->{header[i]}\n" for i in range(n)]))
```

```python
#Afficher les 50 premières lignes et les 50 dernières lignes de cette liste.
print(rows[1:50])
```

```python
#Combien d’arbres sont recensés dans ce jeu de données ?
n_arbres = len(rows)-1 #without the header
print(f"Il y a {n_arbres} arbres recensés")
```

```python
#Pour combien d’arbres manque-t-il l’information concernant la date de plantation ? 
#(Vous pouvez essayer de le faire d’au moins deux manières différentes.)
#methode 1: accumulateur
col_date = 19
accu = 0
for r in rows:
    if r[col_date]=='':
        accu +=1
print(f"Méthode 1 (Avec Accumulateur): Il y a {accu} arbres sans années de plantation")
#méthode 2: avec list.count
n_date = ([r[col_date] for r in rows]).count('')
print(f"Méthode 2 (Avec List.count et liste par compréhension): Il y a {n_date} arbres sans années de plantation")
```

```python
#Combien d’arbres ont été plantés l’année de votre naissance ? 
arbres_1995 = ([r[col_date] for r in rows]).count('1995')
print(f"{arbres_1995} arbres on été plantées en 1995 soit {round(arbres_1995/n_arbres*100, 2)}% des arbres recensés")
```

```python
#Quelle est la plus ancienne année de plantation recensée dans ce dataset ? La plus récente ?    
maxi,imax,mini,imin = 0,[],2077,[]
for i in range(1,len(rows)):
    try:
        y = int(rows[i][col_date])
    except ValueError:
        continue   
    y = int(rows[i][col_date])
    if y < mini:
        imin = [i]
        mini = y
    elif y == mini:
        imin.append(i)
    if y > maxi:
        imax = [i]
        maxi = y
    elif y == maxi:
        imax.append(i)
print(f"Les plus vieux arbres plantées datent de {mini}, {len(imin)} arbres plantées cette année là")
print(f"Les plus jeune arbres plantées datent de {maxi}, {len(imax)} arbres plantées cette année là")
```

```python
#Combien d’arbres ont été plantés année par année (ex : 1987 : 771, 1988 : 266, etc…) ?
liste_annee = [int(r[col_date]) for r in rows[1:] if r[col_date]!='']
# date_cardinal = { y:liste_annee.count(y) for y in liste_annee} # short but not efficient
date_cardinal = {} # run 100 times faster
for y in liste_annee:
    if not (y in date_cardinal.keys()):
        date_cardinal[y]=liste_annee.count(y)
for y in sorted(date_cardinal.keys()):
    print(f"Année {y}:{date_cardinal[y]}")    
```

```python
#Combien d’arbres ont été plantés en moyenne chaque année ?   Moyenne d’une liste de deux manières :
#méthode 1: using sum
non_zero_cardinal = list(date_cardinal.values())
moyenne1 = (sum(non_zero_cardinal)/(maxi-mini+1))
print(f"méthode sum : de {mini} à {maxi}, il y a en moyenne {round(moyenne1,2)} arbres plantées chaque année")
#méthode 2: statistics.mean() or
cardinal = []
for y in range(mini,maxi+1):
    try:
        cardinal.append(date_cardinal[y])
    except KeyError:
        cardinal.append(0)
import statistics
moyenne2 = statistics.mean(cardinal)
print(f"méthode statiststics : de {mini} à {maxi}, il y a en moyenne {round(moyenne2,2)} arbres plantées chaque année")
#méthode 3:Using 
import numpy
moyenne3 = numpy.average(cardinal)
print(f"méthode numpy : de {mini} à {maxi}, il y a en moyenne {round(moyenne2,2)} arbres plantées chaque année")
```

```python
#Stocker conjointement l’année de plantation et le nombre d’arbres plantés dans un tuple. 
#Les tuples seront stockés dans une liste (ex : [ (‘1987’, 771), (‘1988’, 266), ….. ] )
lac = []
for y in date_cardinal.keys():
    lac.append((y,date_cardinal[y]))
print(f"lac = {lac}")


```

```python
#Quel Maire a planté le plus d’arbres à Grenoble ?
def anneeToMaire(y:int):
    if y<=1950:
        return "Léon Martin"
    elif y<=1965:
        return "Albert Michallon"
    elif y <= 1983:
        return "Hubert Dubedout"
    elif y <= 1994:
        return "Alain Carignon"
    elif y <= 2014:
        return "Michel Destot"
    else:
        return "Éric Piolle"
liste_des_maires = ["Léon Martin","Albert Michallon","Hubert Dubedout","Alain Carignon","Michel Destot","Éric Piolle"]
maireToArbre = {"Léon Martin":0,"Albert Michallon":0,"Hubert Dubedout":0,"Alain Carignon":0,"Michel Destot":0,"Éric Piolle":0}
for (y,nby) in lac:
    maireToArbre[anneeToMaire(y)]+=nby
print("Voici le palmarès des maires de Grenoble ayants la main verte ! \n(Roulement de tambour ...)")
for nb in sorted(maireToArbre.values(),reverse=True):
    for maire in liste_des_maires:
        if maireToArbre[maire] == nb:
            print(f"Pendant l´investiture de {maire}, {nb} arbres on été plantées")
```
