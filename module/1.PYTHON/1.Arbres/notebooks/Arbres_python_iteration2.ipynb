```python
import csv
with open('../data/trees.csv', newline='',encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    data = list(reader)

```

```python
#Pour commencer, identifiez tous les genres botaniques représentés à Grenoble. 
#Combien y en a-t-il ?  Essayez plusieurs manières de résoudre ce problème. 
header = data[0]
i_to_n = {} #indexToName
n_to_i = {} #nameToIndex
for i in range(len(header)):
    i_to_n[i] = header[i]
    n_to_i[header[i]]=i
print(f"index_to_name={i_to_n}\n")
print(f"name_to_index={n_to_i}\n")
index_bota = 13
index_espece = 14
genre_bota_list = [r[index_bota] for r in data[1:]]
#méthode 1
genre_bota = set(genre_bota_list)
genre_bota.remove('')
print(f"méthode set: genre bota = {genre_bota}\n")

#méthode 2
genre_bota2=[]
for r in data[1:]:
    g = r[index_bota]
    if g != "" and g not in genre_bota2:
        genre_bota2.append(g)
print(f"méthode (for + list): genre bota = {genre_bota2}\n")
```

```python
#Identifiez maintenant toutes les espèces différentes présentes à Grenoble. 
#N’oubliez pas que les espèces sont une sous-catégorie de genre botanique ! 
#(l’espèce est ‘Platanus Acerifolia’ et pas uniquement ‘Acerifolia’) 
print(f"On calcule l´ensemble  des couples (genre_bota,espece), tel que espece!='' and genre_bota!=''")
espece_list = [(r[index_bota],r[index_espece]) for r in data[1:] if r[index_bota] !='' and r[index_espece]!='']
espece = set(espece_list)
print(espece)
```

```python
#En utilisant la structure de données Dict compter le nombre de représentants de chaque genre botanique : 
# genre_census = {}
# temp = [g]
# for g in genre_bota:
#     genre_census[g] = genre_bota_list.count(g)
def census(l):
    """
    return the dictionarry value:cardinal from a list with repetition
    """
    return {v:l.count(v) for v in set(l) if v!=''}
genre_census = census(genre_bota_list)     
print(genre_census)
```

```python
#Quel est le genre botanique le plus représenté ?
def clef_majoritaire(dico):
    return max(dico,key=dico.get)
genre_majoritaire = max(genre_census,key=genre_census.get)
print(f"Le genre le plus fréquent est {genre_majoritaire} avec {genre_census[genre_majoritaire]} représentants ")
```

```python
#Listez tous les représentants uniques de leurs genres botaniques  
unique_rep = [g for g in genre_census.keys() if genre_census[g]==1]
print(f"Les représentants unique de leurs genre sont:\n{unique_rep}")
```

```python
#Classer tous les genres botaniques en fonction de leur nombre (ordre décroissant)
genre_sorted = sorted(list(genre_bota),key=genre_census.get,reverse=True)
print(f"Voici la liste des genre par ordre decroissant de leur nombre\n{genre_sorted}")
```

```python
#Quelle est la proportion du genre botanique le plus représenté parmi tous les genres ? 
def proportion(k,dico:dict):
    return round(dico[k]/sum(dico.values())*100,2)
    
def census_proportion(dico:dict):
    return {k:(proportion(k,dico)) for k in dico.keys()}
prop=proportion(genre_majoritaire,genre_census)
print(f"Le genre le plus fréquent {genre_majoritaire}, représente {prop}% de la population d'arbres dont le genre est connu")
```

```python
#Quelle est la proportion de chacun des genres botaniques par rapport au total ? 
genre_proportion = census_proportion(genre_census)
print(f"voici le dico, des genres associé á leur proportions (en %)\n{genre_proportion}")
```

```python
#Les objectifs de la Métro sont-ils respectés en termes de diversité ?


# objectif 1: pas plus de 10% d'une espece;
espece_census = census(espece_list)
espece_proportion = census_proportion(espece_census)
espece_majoritaire = clef_majoritaire(espece_census)
print(f"l´espece la plus répandue, le {espece_majoritaire}, est présente a {espece_proportion[espece_majoritaire]}% parmis les arbres dont l'espece est repertoriée, le seuil de 10% est donc dépassé")
# objectif 2: pas plus de 20% d´un genre;
print(f"Le genre le plus répandu, le {genre_majoritaire}, est présente a {genre_proportion[genre_majoritaire]}% parmis les arbres dont l'espece est repertoriée, le seuil de 20% n'est donc dépassé")
```

```python
#(Plus dur) Comment la pratique de plantation a-t-elle évolué au cours du temps ? Diversifie-t-on plus maintenant qu’avant? 
#preprocessing
import numpy as np
import matplotlib.pyplot as plt
date_genreEspece_list = []
for r in data[1:]:
    if r[19]!='' and r[13]!='' and r[14]!='':
        date = int(r[19])
        espece = (r[13],r[14])
        date_genreEspece_list.append((date,espece))
#clustering par tranche de 10 ans

# dico = {y:([(a,b) for (a,b) in date_genreEspece_list if a < y+10]) for y in range(2000, 2020,10)}
et_spec_prop = {} # y:ec où ec ecart type de la proportion pour chaque periode [y,y+10[ 
et_spec_pop = {} # y:ec où ec ecart type de la population pour chaque periode [y,y+10[ 
spec_maj_prop = {}
for y in range(1950,2021,10):
    spec_census = census([e for (d,e) in date_genreEspece_list if d < y+10])
    spec_proportion = census_proportion(spec_census)
    spec_maj = clef_majoritaire(spec_census)
    props = list(spec_proportion.values())
    pops = list(spec_census.values())
    et_spec_prop[y]= np.std(props)
    et_spec_pop[y]= np.std(pops)
    spec_maj_prop[y] = spec_proportion[spec_maj]

#Observation 1
print("On va regarder l'évolution, de l'écart-type de la population de chaque espece (en nombre d'arbre), par tranche de 10 ans depuis 1950")
print(et_spec_pop)
dico = et_spec_pop
xvalues = list(dico.keys())
yvalues = list(dico.values())
plt.plot(xvalues, yvalues,marker = 'o')
plt.xlabel("Années")
plt.ylabel("EC(population des espèces)")
plt.show() 
#conclusion
print("conclusion: Pas très concluant, l´écart-type augmente (avec le nombre d'arbres plantées )\n"+"-"*141)

#Observation 2
print("On va regarder l'évolution, de l'écart-type de la proportion d'espece (en %), par tranche de 10 ans depuis 1950")
print(et_spec_prop)
dico = et_spec_prop
xvalues = list(dico.keys())
yvalues = list(dico.values())
plt.plot(xvalues, yvalues,marker = 'o')
plt.xlabel("Années")
plt.ylabel("EC(proportion des espèces)")
plt.show() 
#conclusion
print("conclusion: Il y a plus de diversité, les espèces sont plantées de manière plus égalitaire au fil des années\n"+"-"*141)

#Observation 3
print("On va regarder l'évolution, de la proportion (en %) de l'espèce majoritaire plantée par tranche de 10 ans depuis 1950")
print(spec_maj_prop)
dico = spec_maj_prop
xvalues = list(dico.keys())
yvalues = list(dico.values())
plt.plot(xvalues, yvalues,marker = 'o')
plt.xlabel("Années")
plt.ylabel("proportion de espèces majoritaire plantée")
plt.show() 
#conclusion
print("conclusion: Supporte les conclusions 2, les espèces les plus plantées sont de moins en moins majoritaire\n"+"-"*141)
```
