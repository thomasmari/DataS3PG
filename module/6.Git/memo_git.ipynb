## Code Versionning


- add files 
```bash
(base) tmari@PYTHON-02:~/Documents/module/6.Git$7,10:22$git add memo_git.ipynb 
...
```
- Commit staged changes with a message
```bash
git commit -m "message"
```
- Commit all tracked changes (skip staging)
```bash
git commit -a -m "message" 
```
- revert uncommited changes
```bash
(base) tmari@PYTHON-02:~/Documents/module/6.Git$36,11:12$git restore temp.txt 
```
- unstage all files you might have staged with git add:

```bash
git reset
```
- revert all local uncommitted changes (should be executed in repo root):
```bash
git checkout . # for all files
git checkout <comit/branchID> # from a specific branch or commit
```
- revert uncommitted changes only to particular file or directory:
```bash 
git checkout [some_dir|file.txt]
```
- revert all uncommitted changes (longer to type, but works from any subdirectory):
```bash
git reset --hard HEAD
```
- remove all local untracked files, so only git tracked files remain:        WARNING: -x will also remove all ignored files, including ones specified by .gitignore! You may want to use -n for preview of files to be deleted.
```bash
git clean -fdx
```
- reset the world: equivalent to fresh git clone from original source (but it does not re-download anything, so is much faster):
```bash
git reset
git checkout .
git clean -fdx
```


## Urgency stash
- Stash your changes
```bash
git stash 
git stash push -m "message" # with a message
```
```bash
```
- List all stashes
```bash
git stash list 
git stash show #See what was changed in the latest stash
```
- Create a branch from a stash
```bash
git stash branch <branchname> 
```
- Restore a stash:
```bash
git stash pop # The Best :Pop (apply and remove) the Stash 
git stash apply # apply the most recent stash
git stash apply stash@{1} # for a specific stash
```
- Remove a stash in case you didnt use pop:
```bash
git stash drop # just remove
git stash clear #remove all stash
```


## Reviewing changes
```bash
git diff # see unstaged changes
git diff <commit1> <commit2> # Compare Two Commits
git log -oneline #- See concise commit history
git log --stat # history per files
```
